```python
#import dependencies & setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
%matplotlib inline
```

```python
# Load CSV files
may_2023_data_to_load = Path("Resources/JC-202305-citibike-tripdata.csv")
june_2023_data_to_load = Path("Resources/JC-202306-citibike-tripdata.csv")
july_2023_data_to_load = Path("Resources/JC-202307-citibike-tripdata.csv")

# read CSV
data_may_2023 = pd.read_csv(may_2023_data_to_load)
data_june_2023 = pd.read_csv(june_2023_data_to_load)
data_july_2023 = pd.read_csv(july_2023_data_to_load)


# convert to pandas df
citibike_may_2023_df = pd.DataFrame(data_may_2023)
citibike_june_2023_df = pd.DataFrame(data_june_2023)
citibike_july_2023_df = pd.DataFrame(data_july_2023)
```

```python
# 1. How many trips have been recorded during the chosen period(s)?

# May 
may_trips = citibike_may_2023_df['ride_id'].nunique()
print(f'There were a total of {may_trips} Citi Bike trips recorded during May 2023.')

# June
june_trips = citibike_june_2023_df['ride_id'].nunique()
print(f'There were a total of {june_trips} Citi Bike trips recorded during June 2023.')

# July
july_trips = citibike_july_2023_df['ride_id'].nunique()
print(f'There were a total of {july_trips} Citi Bike trips recorded during July 2023.')
```

```python
# Concatenate/merge all three dfs into one master
frames = [citibike_may_2023_df, citibike_june_2023_df, citibike_july_2023_df]

result = pd.concat(frames)
result.head()
```
